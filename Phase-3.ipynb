## Statistical and Categorical Plots


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(42)
data = np.random.normal(loc=60, scale=12, size=500)

plt.figure(figsize=(8, 5))
plt.hist(data, bins=20, edgecolor='black')
plt.title('Histogram - Exam Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

## Boxplot and Violin-like comparison


```python
group_a = np.random.normal(65, 8, 100)
group_b = np.random.normal(72, 10, 100)
group_c = np.random.normal(78, 7, 100)

plt.figure(figsize=(8, 5))
plt.boxplot([group_a, group_b, group_c], labels=['A', 'B', 'C'])
plt.title('Boxplot - Group Score Distribution')
plt.xlabel('Group')
plt.ylabel('Score')
plt.grid(axis='y', alpha=0.25)
plt.show()
```

## Pie and Stacked Bar


```python
categories = ['Rent', 'Food', 'Travel', 'Savings']
values = [35, 25, 15, 25]

plt.figure(figsize=(7, 7))
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90)
plt.title('Monthly Budget Split')
plt.show()

months = ['Jan', 'Feb', 'Mar', 'Apr']
product_a = [20, 25, 30, 35]
product_b = [15, 18, 20, 23]

plt.figure(figsize=(8, 5))
plt.bar(months, product_a, label='Product A')
plt.bar(months, product_b, bottom=product_a, label='Product B')
plt.title('Stacked Bar Plot')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

## Error Bars


```python
x = np.arange(1, 6)
mean_values = [20, 24, 27, 26, 30]
errors = [1.5, 2.0, 1.2, 1.8, 1.3]

plt.figure(figsize=(8, 5))
plt.errorbar(x, mean_values, yerr=errors, fmt='-o', capsize=5)
plt.title('Error Bar Plot')
plt.xlabel('Experiment')
plt.ylabel('Measured Value')
plt.grid(alpha=0.3)
plt.show()
```
