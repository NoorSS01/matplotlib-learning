## Business Dashboard Style with Matplotlib


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
sales_df = pd.DataFrame(
    {
        'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
        'north': [120, 135, 150, 160, 172, 180],
        'south': [110, 125, 140, 150, 165, 170],
        'west': [100, 118, 130, 145, 155, 162],
        'online_orders': [220, 250, 275, 300, 335, 360],
    }
)

print(sales_df)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

axes[0, 0].plot(sales_df['month'], sales_df['north'], marker='o', label='North')
axes[0, 0].plot(sales_df['month'], sales_df['south'], marker='o', label='South')
axes[0, 0].set_title('Region Trend')
axes[0, 0].legend()
axes[0, 0].grid(alpha=0.25)

axes[0, 1].bar(sales_df['month'], sales_df['west'], color='slateblue')
axes[0, 1].set_title('West Monthly Sales')
axes[0, 1].grid(axis='y', alpha=0.25)

axes[1, 0].scatter(sales_df['north'], sales_df['online_orders'], color='darkgreen')
axes[1, 0].set_title('North vs Online Orders')
axes[1, 0].set_xlabel('North Sales')
axes[1, 0].set_ylabel('Online Orders')
axes[1, 0].grid(alpha=0.25)

share = [sales_df['north'].sum(), sales_df['south'].sum(), sales_df['west'].sum()]
axes[1, 1].pie(share, labels=['North', 'South', 'West'], autopct='%1.1f%%', startangle=90)
axes[1, 1].set_title('Regional Share')

plt.tight_layout()
plt.show()
```

```python
sales_df['total_regions'] = sales_df[['north', 'south', 'west']].sum(axis=1)
rolling_avg = sales_df['total_regions'].rolling(3).mean()

plt.figure(figsize=(9, 5))
plt.plot(sales_df['month'], sales_df['total_regions'], marker='o', label='Total Regional Sales')
plt.plot(sales_df['month'], rolling_avg, marker='s', label='3-Month Rolling Avg')
plt.title('Total Sales and Rolling Average')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
