## Matplotlib Styling and Customization


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 200)

plt.figure(figsize=(9, 5))
plt.plot(x, np.sin(x), color='navy', linestyle='-', linewidth=2, marker='o', markevery=20, label='sin(x)')
plt.plot(x, np.cos(x), color='darkorange', linestyle='--', linewidth=2, label='cos(x)')
plt.title('Line Styles, Colors and Markers', fontsize=14)
plt.xlabel('x')
plt.ylabel('value')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

## Subplots


```python
x = np.arange(1, 11)

fig, axes = plt.subplots(2, 2, figsize=(10, 7))
axes[0, 0].plot(x, x, color='blue')
axes[0, 0].set_title('Linear')

axes[0, 1].plot(x, x ** 2, color='green')
axes[0, 1].set_title('Quadratic')

axes[1, 0].plot(x, np.sqrt(x), color='purple')
axes[1, 0].set_title('Square Root')

axes[1, 1].plot(x, np.log(x), color='red')
axes[1, 1].set_title('Log')

for ax in axes.flat:
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.grid(alpha=0.2)

plt.tight_layout()
plt.show()
```

## Annotations and Text


```python
x = np.arange(1, 8)
y = [20, 35, 30, 45, 50, 48, 60]

plt.figure(figsize=(9, 5))
plt.plot(x, y, marker='o')
max_idx = int(np.argmax(y))

plt.annotate('Highest point', xy=(x[max_idx], y[max_idx]), xytext=(x[max_idx]-1, y[max_idx]-10),
             arrowprops=dict(arrowstyle='->', color='black'))

plt.text(1, 58, 'Weekly trend', fontsize=11)
plt.title('Annotations Example')
plt.xlabel('Day')
plt.ylabel('Units Sold')
plt.grid(alpha=0.3)
plt.show()
```

## Styles


```python
print('Available style sample:', plt.style.available[:8])

plt.style.use('ggplot')
x = np.linspace(0, 2 * np.pi, 150)

plt.figure(figsize=(8, 5))
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.title('ggplot Style Example')
plt.legend()
plt.show()

plt.style.use('default')
```
