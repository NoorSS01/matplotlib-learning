## Matplotlib Basics and First Plots


```python
import matplotlib.pyplot as plt
import numpy as np
```

Create first line plot


```python
x = np.arange(0, 10)
y = x ** 2

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title('Basic Line Plot')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid(True)
plt.show()
```

### Plotting multiple lines


```python
x = np.arange(1, 11)
line_1 = x
line_2 = x * 2
line_3 = x ** 1.5

plt.figure(figsize=(8, 5))
plt.plot(x, line_1, label='y = x')
plt.plot(x, line_2, label='y = 2x')
plt.plot(x, line_3, label='y = x^1.5')
plt.title('Multiple Line Plot')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
plt.grid(True)
plt.show()
```

### Bar and Scatter Plot


```python
students = ['A', 'B', 'C', 'D', 'E']
marks = [72, 85, 68, 91, 77]

plt.figure(figsize=(8, 5))
plt.bar(students, marks)
plt.title('Student Marks - Bar Plot')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.show()

hours = [1, 2, 3, 4, 5, 6, 7]
scores = [40, 45, 52, 57, 65, 71, 80]

plt.figure(figsize=(8, 5))
plt.scatter(hours, scores)
plt.title('Study Hours vs Score - Scatter Plot')
plt.xlabel('Study Hours')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

### Figure and Axis basics


```python
fig, ax = plt.subplots(figsize=(8, 5))
x = np.linspace(0, 2 * np.pi, 100)
ax.plot(x, np.sin(x), color='teal')
ax.set_title('Sine Wave using Figure and Axis')
ax.set_xlabel('Angle')
ax.set_ylabel('sin(x)')
ax.grid(True)
plt.show()
```
