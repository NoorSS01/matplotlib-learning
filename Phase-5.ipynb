## Saving Figures and Plot Export


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2 * np.pi, 300)

plt.figure(figsize=(10, 5))
plt.plot(x, np.sin(x), label='sin(x)', linewidth=2)
plt.plot(x, np.cos(x), label='cos(x)', linewidth=2)
plt.title('Sine and Cosine Waves')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(alpha=0.3)
plt.savefig('trig-plot.png', dpi=200, bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
products = ['A', 'B', 'C', 'D']
q1_sales = [40, 55, 30, 60]

ax.bar(products, q1_sales, color='coral')
ax.set_title('Q1 Product Sales')
ax.set_xlabel('Product')
ax.set_ylabel('Sales')
fig.savefig('q1-sales.pdf', bbox_inches='tight')
plt.show()
```

```python
try:
    image = plt.imread('trig-plot.png')
    plt.figure(figsize=(9, 4))
    plt.imshow(image)
    plt.title('Loaded Saved Plot Image')
    plt.axis('off')
    plt.show()
except FileNotFoundError:
    print('Saved plot file not found')
```
